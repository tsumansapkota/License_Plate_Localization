```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import os
%matplotlib inline
```

### Loading the json data file

```python
# GUYS = "dactar" 
# GUYS = "ashish" 
# GUYS = "jeevan"
GUYS = "raju"
JSON_FILE = f'data/data_{GUYS}.json'

image_points0 = {}
with open(JSON_FILE, "r") as dataFile:
    import copy
    data = dataFile.read()
    if data is not None and data != '':
        jsonData = json.loads(data)
        image_points0 = copy.deepcopy(
            {**jsonData, **image_points0})
```

### Preprocessing the points

```python
def get_fourth_point(row):
    p30 = row['p0'][0] - (row['p1'][0] - row['p2'][0])
    p31 = row['p0'][1] - (row['p1'][1] - row['p2'][1])
    return [p30, p31]
# xdata=[cps[0][0], cps[1][0], event.xdata, cps[0][0]-(cps[1][0] - event.xdata),cps[0][0]]
# ydata=[cps[0][1], cps[1][1], event.ydata, cps[0][1]-(cps[1][1] - event.ydata),cps[0][1]]
```

```python
df0['p3'] = df0.apply(get_fourth_point, axis=1)
# df['classification'] = df.apply(classify_example, axis = 1, args=(tree,))
# df['classification_correct'] = df.classification == df.label
```

### Managing the order of points for Learning

```python
p0 = np.array(list(df0.p0))
p1 = np.array(list(df0.p1))
p2 = np.array(list(df0.p2))
p3 = np.array(list(df0.p3))
```

```python
midpoint = (p0+p1+p2+p3)/4
midpoint
```

```python
(p1<midpoint).sum(axis=0)
```

```python
for i in range(len(midpoint)):
    if p1[i][0] < midpoint[i][0]:
        print(i, midpoint[i])
        print(df0.iloc[i])
        print()
```

```python
sorted_pts = []
points = [p0,p1,p2,p3]
for i in range(len(midpoint)):
    leftpts = []
    rightpts = []
    sortedpts = [0,0,0,0]
    for pt in points:
        if pt[i][0] < midpoint[i][0]:
            leftpts.append(pt[i])
        else:
            rightpts.append(pt[i])
    if leftpts[0][1] < leftpts[1][1]:
        sortedpts[0] = leftpts[0]
        sortedpts[3] = leftpts[1]
    else:
        sortedpts[0] = leftpts[1]
        sortedpts[3] = leftpts[0]
    if rightpts[0][1] < rightpts[1][1]:
        sortedpts[1] = rightpts[0]
        sortedpts[2] = rightpts[1]
    else:
        sortedpts[1] = leftpts[1]
        sortedpts[2] = leftpts[0]
    sorted_pts.append(sortedpts)
sorted_pts = np.array(sorted_pts)
```

```python
sorted_pts.shape
```

```python
p0t = sorted_pts[:,0,:]
p1t = sorted_pts[:,1,:]
p2t = sorted_pts[:,2,:]
p3t = sorted_pts[:,3,:]
```

### Fitting all images into the dataset

```python
df0['labeled'] = [1 for _ in range(len(df0))]
df0.head()
```

```python
labeled_img0 = df0.image.values.tolist()
labeled_img0
```

```python
count = 0
for lblimg in labeled_img0:
    if lblimg not in all_images0:
        count +=1
        print(count, lblimg, end=' ')
```

```python
# all_images0 - labeled_img0
unlabeled_img0 = []
for allimg in all_images0:
    if allimg not in labeled_img0:
        unlabeled_img0.append(allimg)
unlabeled_img0
```

```python
df0.points[1]
```

```python
for new_im in unlabeled_img0:
    newdf = pd.DataFrame(data={'image':new_im,
                               'points':[np.zeros_like(df0.points[0])],
                               'labeled':0})
    df0 = df0.append(newdf, ignore_index=True)
```

```python
df0
```

```python
# renaming the images which were duplicated
imgnames = df0.image.tolist()
for i in range(len(imgnames)):
    if "(1)" in imgnames[i]:
        imgnames[i] = imgnames[i].replace(" ", "")
        imgnames[i] = imgnames[i].replace("(1)", "")

df0 = df0.drop(columns=['image'])
df0['image'] = imgnames
df0 = df0[['image', 'points', 'labeled']]
df0.info()
```
